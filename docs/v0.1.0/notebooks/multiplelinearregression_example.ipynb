# MultipleLinearRegression Example
Interactive playground for LineSight `MultipleLinearRegression`. This notebook contains every explanatory and visualization method available for this model.

```python
!pip install linesight
import numpy as np
import matplotlib.pyplot as plt
from linesight import MultipleLinearRegression
# Generate sample data
X = np.random.randn(100, 3)
y = X @ [2.5, -1.0, 0.5] + 1.0 + np.random.randn(100) * 0.5

model = MultipleLinearRegression(store_history=True)
model.fit(X, y)
```

### `.summary()`

```python
model.summary()
```

### `.plot_fit()`

```python
model.plot_fit(X, y)
```

### `.plot_prediction_plane()`

```python
model.plot_prediction_plane(X, y)
```

### `.plot_partial_regression()`

```python
model.plot_partial_regression(X, y, feature_idx=0)
```

### `.plot_feature_importance()`

```python
model.plot_feature_importance()
```

### `.plot_correlation_matrix()`

```python
model.plot_correlation_matrix(X, y)
```

### `.plot_multicollinearity()`

```python
model.plot_multicollinearity(X)
```

### `.plot_residual_heatmap()`

```python
model.plot_residual_heatmap(X, y)
```
