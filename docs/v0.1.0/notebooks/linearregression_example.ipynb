# LinearRegression Example
Interactive playground for LineSight `LinearRegression`. This notebook contains every explanatory and visualization method available for this model.

```python
!pip install linesight
import numpy as np
import matplotlib.pyplot as plt
from linesight import LinearRegression
# Generate sample data
X = np.linspace(0, 10, 100).reshape(-1, 1)
y = 2.5 * X.ravel() + 1.0 + np.random.randn(100) * 0.5

model = LinearRegression(store_history=True)
model.fit(X, y)
```

### `.show_equation()`

```python
model.show_equation()
```

### `.explain_coefficients()`

```python
model.explain_coefficients()
```

### `.explain_fit()`

```python
model.explain_fit(X, y)
```

### `.plot_fit()`

```python
model.plot_fit(X, y)
```

### `.plot_residuals()`

```python
model.plot_residuals(X, y)
```

### `.plot_loss_curve()`

```python
model.plot_loss_curve()
```

### `.animate_training()`

```python
model.animate_training(X, y)
```

### `.plot_loss_surface()`

```python
model.plot_loss_surface(X, y)
```

### `.plot_gradient_vectors()`

```python
model.plot_gradient_vectors(X, y)
```

### `.plot_prediction_intervals()`

```python
model.plot_prediction_intervals(X, y)
```

### `.plot_sensitivity_analysis()`

```python
model.plot_sensitivity_analysis(X, y)
```
