# LineSight — Complete Interactive Tutorial

This notebook exercises **every single public function** across all 7 regression models in LineSight.

| Model | Type |
|---|---|
| `LinearRegression` | Single-feature gradient descent |
| `MultipleLinearRegression` | Multi-feature gradient descent |
| `PolynomialRegression` | Non-linear feature expansion |
| `RidgeRegression` | L2 regularization |
| `LassoRegression` | L1 regularization + feature selection |
| `ElasticNetRegression` | L1 + L2 combined |
| `LogisticRegression` | Binary classification |

> **After every diagram**, a **Context Panel** is printed to stdout explaining the theory, the formula, the constraints, and how to read the chart.

## Setup — imports and shared datasets

```
import numpy as np
import warnings
warnings.filterwarnings("ignore")      # suppress convergence hints in output
from IPython.display import HTML       # needed to display JS animations

from linesight import (
    LinearRegression,
    MultipleLinearRegression,
    PolynomialRegression,
    RidgeRegression,
    LassoRegression,
    ElasticNetRegression,
    LogisticRegression,
)

np.random.seed(42)

# ── 1-D dataset (for LinearRegression / PolynomialRegression) ──────────────
X_lin = np.random.rand(100, 1) * 10
y_lin = 2.5 * X_lin.ravel() + 5.0 + np.random.randn(100) * 2.0

# ── 3-feature dataset (for Multiple / Ridge / Lasso / ElasticNet) ──────────
X_mult = np.random.rand(100, 3) * 10
y_mult = (3.0 * X_mult[:, 0]
          - 1.5 * X_mult[:, 1]
          + 2.0 * X_mult[:, 2]
          + 10.0
          + np.random.randn(100))

# ── 2-feature dataset (for constraint region plots that need exactly 2 cols) ─
X_2f   = X_mult[:, :2]          # first 2 columns
y_2f   = (3.0 * X_2f[:, 0]
          - 1.5 * X_2f[:, 1]
          + 10.0
          + np.random.randn(100))

# ── Curved 1-D dataset (for PolynomialRegression) ──────────────────────────
X_poly = np.linspace(-3, 3, 120).reshape(-1, 1)
y_poly = (0.5 * X_poly.ravel()**3
          - X_poly.ravel()**2
          + 2.0
          + np.random.randn(120) * 1.5)

# ── Binary classification dataset (for LogisticRegression) ─────────────────
X_log  = np.random.randn(200, 2) * 2
z_log  = 1.5 * X_log[:, 0] - 2.0 * X_log[:, 1] + 0.5
y_log  = (np.random.rand(200) < (1 / (1 + np.exp(-z_log)))).astype(float)

print("Datasets ready.")
```

---
## 1  Simple Linear Regression

**Concept:** Learns `y = m*x + b` via gradient descent, minimising MSE.

**All functions covered:**
`fit` · `predict` · `score` · `get_training_history` · `show_equation` · `explain_coefficients` · `explain_fit` · `summary` · `plot_fit` · `plot_residuals` · `plot_loss_curve` · `plot_loss_surface` · `plot_gradient_vectors` · `plot_prediction_intervals` · `plot_sensitivity_analysis` · `plot_actual_vs_predicted` · `plot_learning_curve` · `compare_learning_rates` · `animate_training` · `animate_loss_surface_path`

```
model_lin = LinearRegression(learning_rate=0.01, epochs=500, store_history=True)
model_lin.fit(X_lin, y_lin)
```

### 1.1  Textual Explanations

```
model_lin.summary()
```

```
model_lin.show_equation()
```

```
model_lin.explain_coefficients()
```

```
model_lin.explain_fit(X_lin, y_lin)
```

### 1.2  Prediction & Scoring

```
preds = model_lin.predict(X_lin[:5])
print("First 5 predictions:", preds)
print()
metrics = model_lin.score(X_lin, y_lin)
print("Score metrics:", metrics)
```

```
history = model_lin.get_training_history()
print(f"Epochs run   : {history.epochs_run}")
print(f"Converged    : {history.converged}")
print(f"Final loss   : {history.losses[-1]:.6f}")
```

```
model_lin.plot_loss_surface(X_lin, y_lin)
```

```
model_lin.compare_learning_rates(X_lin, y_lin, learning_rates=[0.001, 0.01, 0.05])
```
