# LogisticRegression Example
Interactive playground for LineSight `LogisticRegression`. This notebook contains every explanatory and visualization method available for this model.

```python
!pip install linesight
import numpy as np
import matplotlib.pyplot as plt
from linesight import LogisticRegression
# Generate sample data
X = np.random.randn(100, 2)
y = (X @ [2.5, -1.0] + 1.0 > 0).astype(int)

model = LogisticRegression(store_history=True)
model.fit(X, y)
```

### `.summary()`

```python
model.summary()
```

### `.explain_boundary()`

```python
model.explain_boundary()
```

### `.plot_sigmoid()`

```python
model.plot_sigmoid(X, y)
```

### `.plot_decision_boundary()`

```python
model.plot_decision_boundary(X, y)
```

### `.plot_probability_surface()`

```python
model.plot_probability_surface(X, y)
```

### `.plot_confusion_matrix()`

```python
model.plot_confusion_matrix(X, y)
```

### `.plot_roc_curve()`

```python
model.plot_roc_curve(X, y)
```

### `.plot_calibration_curve()`

```python
model.plot_calibration_curve(X, y)
```

### `.plot_threshold_sensitivity()`

```python
model.plot_threshold_sensitivity(X, y)
```

### `.plot_log_odds()`

```python
model.plot_log_odds(X, y)
```

### `.animate_boundary()`

```python
model.animate_boundary(X, y)
```

### `.plot_residuals()`

```python
model.plot_residuals(X, y)
```
