# LassoRegression Example
Interactive playground for LineSight `LassoRegression`. This notebook contains every explanatory and visualization method available for this model.

```python
!pip install linesight
import numpy as np
import matplotlib.pyplot as plt
from linesight import LassoRegression
# Generate sample data
X = np.random.randn(100, 2)
y = X @ [2.5, 0.0] + 1.0 + np.random.randn(100) * 0.5

model = LassoRegression(alpha=0.5, store_history=True)
model.fit(X, y)
```

### `.summary()`

```python
model.summary()
```

### `.explain_regularization()`

```python
model.explain_regularization()
```

### `.plot_coefficient_shrinkage()`

```python
model.plot_coefficient_shrinkage(X, y)
```

### `.plot_feature_elimination()`

```python
model.plot_feature_elimination(X, y)
```

### `.plot_sparsity_path()`

```python
model.plot_sparsity_path(X, y)
```

### `.plot_constraint_region()`

```python
model.plot_constraint_region(X, y)
```

### `.animate_coordinate_descent()`

```python
model.animate_coordinate_descent(X, y)
```

### `.plot_residuals()`

```python
model.plot_residuals(X, y)
```
