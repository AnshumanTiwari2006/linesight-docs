# RidgeRegression Example
Interactive playground for LineSight `RidgeRegression`. This notebook contains every explanatory and visualization method available for this model.

```python
!pip install linesight
import numpy as np
import matplotlib.pyplot as plt
from linesight import RidgeRegression
# Generate sample data
X = np.random.randn(100, 2)
y = X @ [2.5, -1.0] + 1.0 + np.random.randn(100) * 0.5

model = RidgeRegression(alpha=1.0, store_history=True)
model.fit(X, y)
```

### `.summary()`

```python
model.summary()
```

### `.explain_regularization()`

```python
model.explain_regularization()
```

### `.plot_coefficient_shrinkage()`

```python
model.plot_coefficient_shrinkage(X, y)
```

### `.plot_effective_degrees_of_freedom()`

```python
model.plot_effective_degrees_of_freedom(X, y)
```

### `.plot_bias_variance_tradeoff()`

```python
model.plot_bias_variance_tradeoff(X, y)
```

### `.plot_constraint_region()`

```python
model.plot_constraint_region(X, y)
```

### `.animate_regularization()`

```python
model.animate_regularization(X, y)
```

### `.plot_residuals()`

```python
model.plot_residuals(X, y)
```
