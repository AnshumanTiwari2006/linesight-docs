# ElasticNetRegression Example
Interactive playground for LineSight `ElasticNetRegression`. This notebook contains every explanatory and visualization method available for this model.

```python
!pip install linesight
import numpy as np
import matplotlib.pyplot as plt
from linesight import ElasticNetRegression
# Generate sample data
X = np.random.randn(100, 2)
y = X @ [2.5, 0.0] + 1.0 + np.random.randn(100) * 0.5

model = ElasticNetRegression(alpha=0.5, l1_ratio=0.5, store_history=True)
model.fit(X, y)
```

### `.summary()`

```python
model.summary()
```

### `.explain_regularization()`

```python
model.explain_regularization()
```

### `.plot_coefficient_shrinkage()`

```python
model.plot_coefficient_shrinkage(X, y)
```

### `.plot_l1_l2_balance()`

```python
model.plot_l1_l2_balance(X, y)
```

### `.animate_l1_ratio_sweep()`

```python
model.animate_l1_ratio_sweep(X, y)
```

### `.plot_residuals()`

```python
model.plot_residuals(X, y)
```
